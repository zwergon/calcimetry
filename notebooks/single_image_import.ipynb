# Notebook to upload drill location image file

This note book is to deal with files that do not fit the posabilities I have covered in the python script `clickclick/import_image_mongodb.py`

```python
csvfile = '/work/armitagj/data/csvs/REP4/SUG1101/Photos/imgs2.csv'
csvfile_bkp = '/work/armitagj/data/csvs/REP4/SUG1101/Photos/imgs2_.csv'
shutil.move(csvfile, csvfile_bkp)

input_file = open(csvfile_bkp, 'r')
output_file = open(csvfile, 'w')

for line in input_file:
    line = re.sub(r'{{', '"{{', line)
    line = re.sub(r'}}', '}}"', line)
    line = re.sub(r',,', '', line)
    line = re.sub(r',', ';', line)
    line = re.sub(r'k_Arrow;', 'k_Arrow', line)
    output_file.write(line)

input_file.close()
output_file.close()
df = pd.read_csv(csvfile, sep=';')

df
```

```python
if len(df.columns) == 13:
    print('hello')
    
    df = df.drop(columns=['Path'])

    # convert the coordinates into a list for mongodb
    df['k_Up'] = df['k_Up'].str.replace(r'{', '', regex=True)
    df['k_Up'] = df['k_Up'].str.replace(r'}', '', regex=True)
    df['k_Up'] = df['k_Up'].str.replace(r';', ',', regex=True)

    df['k_Down'] = df['k_Down'].str.replace(r'{', '', regex=True)
    df['k_Down'] = df['k_Down'].str.replace(r'}', '', regex=True)
    df['k_Down'] = df['k_Down'].str.replace(r';', ',', regex=True)

    df['k_Arrow'] = df['k_Arrow'].str.replace(r'{', '', regex=True)
    df['k_Arrow'] = df['k_Arrow'].str.replace(r'}', '', regex=True)
    df['k_Arrow'] = df['k_Arrow'].str.replace(r';', ',', regex=True)

    df['k_Up'] = df['k_Up'].apply(
        lambda x: np.array(x.split(','), dtype=int).reshape(-1, 2)  if(pd.notnull(x)) else x)
    df['k_Down'] = df['k_Down'].apply(
        lambda x: np.array(x.split(','), dtype=int).reshape(-1, 2)  if(pd.notnull(x)) else x)
    df['k_Arrow'] = df['k_Arrow'].apply(
        lambda x: np.array(x.split(','), dtype=int).reshape(-1, 2)  if(pd.notnull(x)) else x)

    # create the json payload
    payload = json.loads(df.to_json(orient='records'))

    # push it to the database if the dirll name is not already
    # there
    with MongoAPI(mongo_info=mongo_info) as mongo_api:
        if mongo_api.db['images'].find_one({'DrillName': df['DrillName'][0]}) is None:
            print('Hello?')
            mongo_api.write_img_many(payload)
```
