```python
import os

for location in DrillLocations:
    path_to_input = '{}{}'.format(filelocs, location)
    dfl = df.loc[df['DrillName'].isin([location])]
    if os.path.exists(path_to_input):
        for fname in os.listdir(path_to_input):
            if os.path.isfile(f'{path_to_input}/{fname}'):
                if '.csv' in fname:
                    # different file names of course
                    if 'Calcimétrie' in fname and 'Zone.Identifier' not in fname:
                        input_file = fname
                    elif 'CALCIMETRIE' in fname and 'Zone.Identifier' not in fname:
                        input_file = fname
                    elif 'Calcimètre' in fname and 'Zone.Identifier' not in fname:
                        input_file = fname
                    elif 'Calci' in fname and 'Zone.Identifier' not in fname:
                        input_file = fname
                    elif 'calcimétrie' in fname and 'Zone.Identifier' not in fname:
                        input_file = fname
        dfcalci = pd.read_csv(f'{path_to_input}/{input_file}')
    
        path_to_image = []
        image_posdown = []
        image_posup = []
        for position in dfcalci['Cote']:
            posdown = dfl['Cote0'].to_numpy()
            posup = dfl['Cote1'].to_numpy()
            position = position * 1000
            for i in range(len(posdown)):
                if position >= float(posdown[i]) and position < float(posup[i]):
                    print(f'{posdown[i]} <= {position} < {posup[i]}')
                    path_to_image.append(dfl['Path'])
                    image_posdown.append(posdown[i])
                    image_posup.append(posup[i])
        
        dfcalci['Path'] = path_to_image
        dfcalci['Cote0'] = image_posdown
        dfcalci['Cote1'] = image_posup      
        
    else:
        print(f'{path_to_input} does not exist')
```
