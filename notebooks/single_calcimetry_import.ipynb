# Notebook to upload drill location calcimetry file

This note book is to deal with files that do not fit the posabilities I have covered in the python script `clickclick/import_calcimetry_mongodb.py`

```python
drill = 'SUG1303'
```

```python
excell2csved = f'/work/armitagj/data/Excell_and_Photos/SUG1303/GTR2005-04_0027_SUG1303_CALCIMETRIE.csv'
df1 = pd.read_csv(excell2csved, sep=',')
df1
```

```python
csvfile = f'/work/armitagj/data/csvs/REP4/{drill}/Photos/imgs2.csv'
csvfile_bkp = f'/work/armitagj/data/csvs/REP4/{drill}/Photos/imgs2_.csv'
shutil.move(csvfile, csvfile_bkp)

input_file = open(csvfile_bkp, 'r')
output_file = open(csvfile, 'w')

for line in input_file:
    line = re.sub(r'{{', '"{{', line)
    line = re.sub(r'}}', '}}"', line)
    line = re.sub(r',,', '', line)
    line = re.sub(r',', ';', line)
    line = re.sub(r'k_Arrow;', 'k_Arrow', line)
    output_file.write(line)

input_file.close()
output_file.close()
df2 = pd.read_csv(csvfile, sep=';')

df2
```

```python
calcifile = f'/work/armitagj/data/csvs/REP4/{drill}/Photos/mesu.csv'

# Modify Renaud's code to create a local mesu.csv
# file, so that it is identical to those already
# created

# I don't have a unique ID for the measurement,
# fix this once the database is complete
id = np.nan

with open(calcifile, 'w') as g:
    g.write("MeasureId;ImageId;CalciCote;CalciVals1m;CalciVals15m\n")
    for i, row2 in df2.iterrows():
        imgId = row2["ImageId"]
        cote0 = row2["Cote0"]
        cote1 = row2["Cote1"]
        for j, row1 in df1.iterrows():
            c = np.rint(row1["Cote"] * 100)
            if cote0 <= c and c <= cote1:
                g.write(f'{id};{imgId};{c};'
                        f'{df1.iloc[j, 2]};'
                        f'{df1.iloc[j, 4]}\n')
    g.close()

```

```python
# read file back in (I know...)
df = pd.read_csv(calcifile, delimiter=';')
df['DrillName'] = drill

df
```

```python
# create the json payload
payload = json.loads(df.to_json(orient='records'))

# push it to the database
with MongoAPI(mongo_info=mongo_info) as mongo_api:
    if mongo_api.db['measurements'].find_one({'DrillName': df['DrillName'][0]}) is None:
        mongo_api.write_mesu_many(payload)
```
